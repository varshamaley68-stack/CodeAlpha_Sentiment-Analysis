```python


import pandas as pd
from textblob import TextBlob

# Sample dataset of reviews
reviews = [
    "I love this product, it works great!",
    "This is the worst experience I’ve ever had.",
    "The service was okay, nothing special.",
    "Absolutely fantastic! Highly recommend.",
    "Not worth the money."
]

# Analyze sentiment
data = []
for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity  # ranges from -1 (negative) to +1 (positive)
    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    data.append([review, sentiment, polarity])

# Create DataFrame
df = pd.DataFrame(data, columns=["Review", "Sentiment", "Polarity"])
print(df)

# Save dataset
df.to_csv("sentiment_analysis.csv", index=False)
```

```python
import matplotlib.pyplot as plt

# Count sentiment categories
sentiment_counts = df["Sentiment"].value_counts()

# Bar chart
plt.figure(figsize=(6,4))
sentiment_counts.plot(kind="bar", color=["green","red","gray"])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

# Pie chart
plt.figure(figsize=(6,6))
sentiment_counts.plot(kind="pie", autopct="%1.1f%%", colors=["green","red","gray"])
plt.title("Sentiment Distribution")
plt.ylabel("")  # Hide y-label
plt.show()
```
